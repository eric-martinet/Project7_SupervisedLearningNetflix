```python
data = pd.read_csv(r"C:\Users\rana_\OneDrive\Bureau\Project7_SupervisedLearningNetflix\data\Netflix.csv")
```

```python
data
```

```python
data.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
data
```

```python
x = data.drop(['category'], axis=1)
y = data['category']
# print(X.head())
print(x.shape)
# print(y.head())
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model = QuadraticDiscriminantAnalysis()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
prediction=model.predict(x_train)

print(confusion_matrix(y_train, prediction))
print(accuracy_score(y_train,prediction))
```

```python
prediction=model.predict(x_test)
print(confusion_matrix(y_test, prediction))
print(accuracy_score(y_test,prediction))
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
```

```python
rfe_selector = RFE(estimator=RandomForestClassifier(),n_features_to_select = 7, step = 1)
m=rfe_selector.fit(x, y)
x.columns[m.get_support()]
print("Num Features: %s" % (m.n_features_))
print("Selected Features: %s" % (m.support_))
print("Feature Ranking: %s" % (m.ranking_))
```

```python

from sklearn.decomposition import PCA
features = x

pca = PCA(n_components=2)
components = pca.fit_transform(x)

```

```python
principal_data = pd.DataFrame(data = components
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principal_data
```

```python
data
```

```python
final_data = pd.concat([principal_data, data[['category']]], axis = 1)
```

```python
final_data
```

```python
x = final_data.drop(['category'], axis=1)
y = final_data['category']
# print(X.head())
print(x.shape)
# print(y.head())
print(y.shape)
```

```python
x_train1, x_test1, y_train1, y_test1 = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train1.shape)
print(y_train1.shape)
print(x_test1.shape)
print(y_test1.shape)
```

```python
model1 = QuadraticDiscriminantAnalysis()
model1.fit(x_train1, y_train1)
prediction=model1.predict(x_train1)

print(confusion_matrix(y_train1, prediction))
print(accuracy_score(y_train1,prediction))
```

```python
prediction=model1.predict(x_test1)
print(confusion_matrix(y_test1, prediction))
print(accuracy_score(y_test1,prediction))
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
categories = ['niche', 'other', 'blockbuster']
colors = ['r', 'g', 'b']
for category, color in zip(categories,colors):
    indicesToKeep = final_data['category'] == category
    ax.scatter(final_data.loc[indicesToKeep, 'principal component 1']
               , final_data.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(categories)
ax.grid()
```
